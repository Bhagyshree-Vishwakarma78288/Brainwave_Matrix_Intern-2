```python
import re
import nltk

# Download NLTK stopwords (if not already installed)
nltk.download('stopwords')
from nltk.corpus import stopwords

# Define function to clean Text
def clean_Text(Text):
    # Convert text to lowercase
    Text = Text.lower()
    
    # Remove URLs
    Text = re.sub(r'http\S+', '', Text)
    
    # Remove mentions and hashtags
    Text = re.sub(r'@\w+', '', Text)
    Text = re.sub(r'#\w+', '', Text)
    
    # Remove special characters and digits
    Text = re.sub(r'[^a-zA-Z\s]', '', Text)
    
    return Text

# Apply text cleaning function
data['cleaned_Text'] = data['Text'].apply(clean_Text)

# Check the cleaned text
data[['Text', 'cleaned_Text']].head()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to get VADER sentiment score
def get_sentiment(Text):
    sentiment = analyzer.polarity_scores(Text)
    return sentiment['compound']  # Compound score (-1 to 1)

# Apply sentiment analysis to cleaned text
data['sentiment_score'] = data['cleaned_Text'].apply(get_sentiment)

# Categorize sentiment based on compound score
data['sentiment_label'] = data['sentiment_score'].apply(lambda x: 'positive' if x > 0.1 else ('negative' if x < -0.1 else 'neutral'))

# Check the sentiment labels
data[['cleaned_Text', 'sentiment_score', 'sentiment_label']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the sentiment labels
sentiment_counts = data['sentiment_label'].value_counts()

# Plot the sentiment distribution
plt.figure(figsize=(8,6))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette='Set2')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
data['Timestamp'] = pd.to_datetime(data['Timestamp'])

# Extract date from timestamp
data['date'] = data['Timestamp'].dt.date

# Group by date and get average sentiment score per day
daily_sentiment = data.groupby('date')['sentiment_score'].mean()

# Plot sentiment trend over time
plt.figure(figsize=(10,6))
daily_sentiment.plot(kind='line', color='purple', marker='o', linewidth=2)
plt.title('Sentiment Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Average Sentiment')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
